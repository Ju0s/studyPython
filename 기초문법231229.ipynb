# 문제
- 학생 성적에 따라 여러분이 등급을 정합니다. 점수는 0에서 100점
- A, B, C, D, F
  + 조건식은 임의로 작성
- 조건문
  + 첫 번째 조건: 90점 이상
  + 두 번째 조건: 75점 이상
  + 세 번째 조건: 60점 이상
  + 네 번째 조건: 45점 이상

```python
x = int(input("점수를 입력하세요..."))
if x >= 90:  # 경계값으로 테스트하는 것이 좋음
  print("A등급")

elif x >= 75:  # 80점 이상~ 90점 미만 범위로 설정하고 싶다면 ==> (x >= 80) and (x < 90)
  print("B등급")

elif x >= 60:
  print("C등급")

elif x >= 45:
  print("D등급")

else:
  print("F등급")
```

```python
# 중첩 조건문
# if 조건문 내부에 또다른 조건문 생성
x = int(input("점수를 입력하세요..."))

# 100점 맞은 사람이 있으면 '장학금' 출력
# 합격을 받은 사람들 중에서 100점인 사람인 경우 장학금을 주겠다
if x >= 90:

  if x == 100:
    print("장학금 수여")

  else:
    print("합격")

else:
  print("불합격")
```

# While 특징
- for-loop 특징: 항상 반복문의 범위를 지정
  + 몇 번 반복문을 수행할지 숫자로 컨트롤 가능
- While 특징
  + 조건문이 False가 될 때까지 반복문 수행
  + 조건문이 True면 반복문은 무한 수행

```python
# 초깃값 지정
i = 1
n = 3

while i <= n:  # 특정 조건문을 False가 되도록 코드를 작성하는 것이 포인트
  print(f'현재 {i}번')
  i = i + 1
```

## for-loop, while 비교

  + 범위가 주어져 있음

## break문
- 중간에 반복문 강제 종료
  + 특정 시점에 반복문 강제 종료

```python
# 초깃값 지정
i = 1
n = 3

# i <= n
while True:  # 특정 조건문을 False가 되도록 코드를 작성하는 것이 포인트
  print(f'현재 {i}번')

  if i >= 5:
     break

  i = i + 1
```

```python
def read_csv(sep, header):
  '''
  Returns the sum of ~~~.

    Parameters:
      sep (str): Character or regex pattern to treat as the delimiter.
      header(int): Row number(s) containing column labels and marking the start of the data (zero-indexed).

  Returns:
    c(int): Sum of ~~
  '''
  c = sep + header
  return c
```

- 코등 문서화를 좋아한다 ==> 네카라쿠배... 갈 확률이 높아짐
- 코드 문서화를 싫어한다 ==> 네카라쿠배... 갈 확률이 낮아짐

# 클래스 상속
- 부모 > 자식
- 대분류 > 중분류 > 소분류
- 동물
  + 동물의 공통적 특징: 살아 움직인다
- 포유류
  + (살아 움직인다) 새끼를 낳는다
- 파충류
  + (살아 움직인다) 알을 낳는다

```python
class Animal:
  # Attribute 정의(=클래스 변수)
  name = ""

  # 공통함수
  def eat(self):
    print("나는 먹을 수 있음")

class Dog(Animal):  # 클래스 상속을 했다  # 동물에서 강아지로 기본 기능 제공
  #name = ""  # 이 코드는 불필요함

  # 새로운 메서드
  def dance(self):
    print(f"{self.name} 강아지가 춤을 춘다")
```

```python
# 객체 생성
a = Dog()
a.name = "코코"
a.eat()  # Animal 클래스를 사용함
```

```python
a.dance()
```

```python
b = Animal()
b.name = "시고르자브종"
b.eat()
```

```python
b.dance()
```

```python
class Person:
  """
  자기 소개하는 Person 클래스.

  '''
  Attributes
  -----------
  name: str
    사람의 이름을 의미
  age: int
    사람의 나이를 의미
  grade: int
     사람의 현재 학년을 의미

  Methods
  --------
  info(comment=""):
    사람의 개인정보를 출력

  """

  def __init__(self, name, age, grade):
    """

    """
    self.name = name
    self.age = age
    self.grade = grade

  def info(self, comment = ""):
    """

    """
    print(f'내 이름은 {self.name}, 내 나이는 {self.age}, 현재 학년은 {self.grade}. {comment}')
```
