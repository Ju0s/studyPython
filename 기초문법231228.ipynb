### why Class? 프로그래밍 관점

```python
result = 0
def add(num):  # 계산기 1대를 보유한 상태
  global result

  result = result + num  # 결괏값(result)에 입력값(num) 더하기
  return result

print(add(3))  # + 5 ==> 8
print(add(5))
```

- 계산기가 1대 더 필요한 상태

```python
result1 = 0
result2 = 0

def add1(num):  # 계산기 1대를 보유한 상태
  global result1

  result1 = result1 + num  # 결괏값(result)에 입력값(num) 더하기
  return result1

def add2(num):  # 계산기 1대를 보유한 상태
  global result2

  result2 = result2 + num  # 결괏값(result)에 입력값(num) 더하기
  return result2


print(add1(3))  # + 5 ==> 8
print(add1(5))
print(add2(3))
print(add2(8))
```

- 위 코드의 문제점: 같은 기능(덧셈)을 수행하는 함수가 중복될 수 있음, 코드의 가독성이 매우 급격하게 떨어지기 시작
- 점프 투 파이썬 클래스 예제

```python
class Calculator:

  def __init__(self):
    self.result = 0  # result 값을 초기화

  # 클래스 메서드
  def add(self, num):
    self.result = self.result + num
    return self.result

cal1 = Calculator()  # 인스턴스화, 계산기 번호 1
cal2 = Calculator()  # 인스턴스화, 계산기 번호 2
cal3 = Calculator()  # 인스턴스화, 계산기 번호 3

# 클래스의 위대함
# 독립적으로 구현되면서, 동시에 실행됨
# 기능은 덧셈 기능
# 계산기 개수를 무한으로 생성 가능

print(cal1.add(3))
print(cal1.add(5))
print(cal2.add(3))
print(cal2.add(100))
```

### 객체 초기화
- '__init__()': 객체 생성과 동시에 속성값을 지정할 수 있다
  + 주로 클래스 선언 시 변수 초기화

```python
class Bicycle():

  # 변수 초기화
  def __init__(self, wheel_size, color):
    self.wheel_size = wheel_size
    self.color = color

  def stop(self):
    print('자전거({0}, {1}): 정지'.format(self.wheel_size, self.color))
```

```python
my_bicycle = Bicycle(26, 'black')
ur_bicycle = Bicycle(30, 'pink')

my_bicycle.stop()
ur_bicycle.stop()
```
